# Check the atmospheric variables

```python
import xarray as xr
```

```python
data_dir = '../outputs/'
```

```python
def plot_all_models2(ds):
    '''Use faceting feature from xarray'''
    cbar_kwargs = {'orientation': 'horizontal','label': 'cm', 'pad': 0.05}
    ds.CorrectedReggrided_ps.isel(time=-1).plot(x='lon', y='lat', col='model', 
                                                    col_wrap=4, robust=True, 
                                                    cbar_kwargs=cbar_kwargs)
```

```python
mip = 'cmip6'
sce = 'ssp245'
var = 'ps'
#ps_ds = pp.read_ps_ds(data_dir, 'cmip6', sce)
hist_ds = xr.open_mfdataset(
    f'{data_dir}/{mip}_{var}_historical/{mip}_{var}_historical_*.nc')
len(hist_ds.model)
```

```python
plot_all_models2(hist_ds)
```
